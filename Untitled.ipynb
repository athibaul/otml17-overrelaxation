```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 20; M = 50
p1 = np.random.randn(N,2)
p2 = np.random.randn(M,2)
c = np.zeros((N,M))

for i in range(N):
    for j in range(M):
        c[i][j] = np.linalg.norm(p1[i,:]-p2[j,:])
        
epsilon = 1e-2
K = np.exp(-c/epsilon)
print(np.shape(K))
K_ = np.transpose(K)

mu = np.random.rand(N)
mu = mu/np.sum(mu)
nu = np.random.rand(M)
nu = nu/np.sum(nu)
```

```python
a = np.ones(N); b = np.ones(M)
err = []
for k in range(100):
    a = mu / K.dot(b)
    b = nu / K_.dot(a)

gamma = np.diag(a).dot(K.dot(np.diag(b)))
print(gamma)
```

```python
fig=plt.figure(figsize=(12,10))


gamma_n = gamma*np.sqrt(N*M)*10
for i in range(N):
    for j in range(M):
        plt.plot([p1[i,0],p2[j,0]],[p1[i,1],p2[j,1]],color='purple',linewidth=gamma_n[i][j])
        
plt.scatter(p1[:,0],p1[:,1],s=mu*1000*N/2,color='red')
plt.scatter(p2[:,0],p2[:,1],s=nu*1000*M/2,color='blue')
plt.axis('equal')
plt.axis('off')
plt.show()
```
