```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50; M = 50
p1 = np.random.randn(N,2).dot(np.diag([2,0.5]))
p2 = np.random.randn(M,2)
c = np.zeros((N,M))

for i in range(N):
    for j in range(M):
        c[i][j] = np.linalg.norm(p1[i,:]-p2[j,:])
        
epsilon = 1e-2
K = np.exp(-c/epsilon)
print(np.shape(K))
K_ = np.transpose(K)

mu = np.random.rand(N)
mu = mu/np.sum(mu)
nu = np.random.rand(M)
nu = nu/np.sum(nu)
```

```python
a = np.ones(N); b = np.ones(M)
err = []
for k in range(100):
    a = mu / K.dot(b)
    b = nu / K_.dot(a)

gamma = np.diag(a).dot(K.dot(np.diag(b)))
print(gamma)
```

```python
fig=plt.figure(figsize=(12,10))


gamma_n = gamma*np.sqrt(N*M)*5
for i in range(N):
    for j in range(M):
        plt.plot([p1[i,0],p2[j,0]],[p1[i,1],p2[j,1]],color='purple',linewidth=gamma_n[i][j])
        
plt.scatter(p1[:,0],p1[:,1],s=mu*500*N/2,color='red')
plt.scatter(p2[:,0],p2[:,1],s=nu*500*N/2,color='blue')
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
N = 100
xx = np.linspace(0,1,N)
mu = 1 + (0.5<xx)*(0.6>xx)
mu = mu/np.sum(mu)
#nu = np.exp(-(xx-0.2)**2*1000) + np.exp(-(xx-0.7)**2*150)
nu = (0.0<=xx)
nu = nu/np.sum(nu)

xxx,yyy = np.meshgrid(xx,xx)
c = np.abs(xxx-yyy)
epsilon = 1e-3
K = np.exp(-c/epsilon)
K_ = np.transpose(K)
print(K)
```

```python
plt.plot(mu)
plt.plot(nu)
plt.show()
```

```python
a = np.ones(N); b = np.ones(N)
errs_init = []
err = 1
while err>1e-12:
    a = mu / K.dot(b)
    b = nu / K_.dot(a)
    err = np.linalg.norm(mu - a * (K.dot(b))) + np.linalg.norm(nu - b * (K_.dot(a)))
    errs_init.append(err)
    
a_true = a; b_true = b;
aibj_true = np.outer(a,b)
print(aibj_true)
gamma_true = np.diag(a).dot(K.dot(np.diag(b)))

def F(aibj):
    return np.sum(K * (aibj_true * np.log(aibj_true / aibj) - aibj_true + aibj), axis=None)

plt.semilogy(errs_init)
eta = (errs_init[-2]-errs_init[-1])/errs_init[-1]
theta_0 = 2/(1+np.sqrt(eta))
print("eta =",eta,"theta0 =",theta_0)
plt.show()
```

```python
def phi(theta,x):
    return x*(1-x**(-theta)) - theta*np.log(x)

thetas = np.linspace(1,theta_0,5000)[::-1]
def Theta(x):
    positives = (phi(thetas,x) > 0)
    i = np.argmax(positives)
    if positives[i]:
        return thetas[i]
    else:
        return 1
    
    
```

```python
a = np.ones(N); b = np.ones(N)
errs = []
ks = [k for k in range(500)]

a_ = np.ones(N); b_ = np.ones(N)
errs_ = []
theta_a = []; theta_b = []
for k in ks:
    a = mu / K.dot(b)
    b = nu / K_.dot(a)
    #err = np.linalg.norm(mu - a * (K.dot(b))) + np.linalg.norm(nu - b * (K_.dot(a)))
    err = F(np.outer(a,b))
    errs.append(err)
    
    a_ratio = mu / (a_ * K.dot(b_))
    theta = Theta(1/np.max(a_ratio))
    theta_a.append(theta)
    a_ = a_*(a_ratio**theta)
    b_ratio = nu / (b_ * K_.dot(a_))
    theta = Theta(1/np.max(b_ratio))
    theta_b.append(theta)
    b_ = b_*(b_ratio**theta)
    #err = np.linalg.norm(mu - a_ * (K.dot(b_))) + np.linalg.norm(nu - b_ * (K_.dot(a_)))
    err = F(np.outer(a_,b_))
    errs_.append(err)

theta_a = np.array(theta_a); theta_b = np.array(theta_b)
#print(np.array(errs))
gamma = np.diag(a).dot(K.dot(np.diag(b)))


plt.style.use(['default','seaborn-poster'])
#plt.imshow(gamma)
fig,ax1 = plt.subplots(figsize=(6,0.618*6))
#ax1.semilogy(ks,errs,label='Sinkhorn')
ax1.semilogy(ks,errs_,label='$F(\\gamma^\\ell)$')
#plt.legend(loc='lower left')
ax1.set_ylabel('Error $F(\\gamma^\\ell)$')
ax1.set_xlabel('Iteration $\ell$')
#ax1.grid(True)
ax2 = ax1.twinx()
ax2.plot([],[])
ax2.plot(ks,theta_a,'--',label='$\\theta_1$')
ax2.plot(ks,theta_b,'--',label='$\\theta_2$')
ax2.grid(True)
ax2.set_ylabel('Overrelaxation $\\theta$')
plt.legend(loc='center right')
plt.savefig('typical_convergence_3.png',transparent=True)
plt.show()
```
